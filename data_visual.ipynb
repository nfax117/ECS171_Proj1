```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
R2 = [0.91857, 0.93629, 0.99228, 0.99201]
MSE = [0.0024572891, 0.0019225386, 0.0002864009,0.0002967862]
```

```python
index = np.arange(4)
width = 0.35
labels = ['Univar_condensed', 'Mulvar_condensed', 'Univar_full', 'Mulvar_full']
# plt.bar(index, R2, width, label='R2')
# plt.bar(index + width, MSE, width,
#     label='MSE')

# plt.ylabel('Scores')
# plt.title('Time series models performance')

# plt.xticks(index + width / 2, ('Univar_condensed', 'Mulvar_condensed', 'Univar_full', 'Mulvar_full'))
# plt.legend(loc='best')

fig, ax = plt.subplots(figsize = (10, 7))
ax2 = ax.twinx()

data1 = ax.bar(index - width/2, R2, width, color='coral', label='R2')
ax.set_ylabel("R2 score", fontsize="large")
ax.bar_label(data1, padding=3)

data2 = ax2.bar(index + width/2, MSE, width, color='skyblue', label='MSE')
ax2.set_ylabel("MSE score", fontsize="large")
ax2.bar_label(data2, padding=3)

ax.set_xlabel("Models", fontsize="large")
ax.set_xticks(index, labels, fontsize="large")
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
fig.tight_layout()
plt.show()
```
